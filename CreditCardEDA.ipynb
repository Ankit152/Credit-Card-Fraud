**The dataset is highly imbalanced.**

```python
plt.figure(figsize=(10,8))
sns.countplot(data['Class'])
plt.title('COUNTPLOT for Class')
plt.xlabel('LABEL')
plt.ylabel('COUNT')
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(data.corr())
plt.title('HEATMAP for the Data')
plt.show()
```

**All the features are some what similar to each other in terms of correlation.**
